```python
# Import scipy libraries for special functions
%pylab inline
from scipy import *
from scipy.special import jn, jn_zeros
# Import 3D libraries and cool colormaps
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Vibrational modes of drumhead
def drumhead_height(n, k, distance, angle, t):
    lam = jn_zeros(n, k) # lamf = kth zero
    lamf = lam[k-1]
    return cos(lamf*t)*cos(n*angle)*jn(n, distance*lamf)

def plot_Bessel(time=0):
    # Lets create the mesh grid based on polar data
    nth, nrad = (75,75) # Resolution
    theta = linspace(0,2*pi,nth)
    radius = linspace(0,1,nrad)
    rm, th = meshgrid(radius,theta)
    xm = rm*cos(th) 
    ym = rm*sin(th)    

    # Lets create a plotting grid with gridspec
    # Create a 3 x 3 plotting grid object with horizontal and vertical spacing wspace and hspace
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2) 
    # Create a figure
    fig = plt.figure(figsize=(13, 6))
    # Define specifications for the plots
    specs = {'rstride' : 1, 'cstride' : 1, 'linewidth' : 0, 'cmap' : cm.jet, 'vmin' : -0.3 , 'vmax' : 0.3}

    # SUBFIGURE 1
    # Create subfigure 1 (takes over three rows  and two columns of the grid)
    n = 2; k = 3
    z1 = drumhead_height(n, k, rm, th, time)
    ax1 = fig.add_subplot(gs[:3, :2], projection='3d')
    s1 = ax1.plot_surface(xm, ym, z1, **specs)
    cbar = fig.colorbar(s1) # Show colorbar
    ax1.axis([-1,1,-1,1])
    ax1.set_zlim3d(-0.6, 0.6) 
    ax1.axis('off')
    title1 = "N=" + str(n) + "  k=" + str(k)
    ax1.set_title('Drum head vibrational modes:\n' + title1, fontsize = 20)

    # SUBFIGURE 2
    # Create subfigure 2 
    n = 2; k = 1
    z2 = drumhead_height(n, k, rm, th, time)
    ax2 = fig.add_subplot(gs[0, 2], projection='3d')
    s2 = ax2.plot_surface(xm, ym, z2, **specs)
    ax2.axis([-1,1,-1,1])
    ax2.set_zlim3d(-0.6, 0.6) 
    ax2.axis('off')
    title2 = "N=" + str(n) + "  k=" + str(k)
    ax2.set_title(title2, fontsize = 20)
    
    # SUBFIGURE 3
    # Create subfigure 3 
    n = 3; k = 2
    z3 = drumhead_height(n, k, rm, th, time)
    ax3 = fig.add_subplot(gs[1, 2], projection='3d')
    s3 = ax3.plot_surface(xm, ym, z3, **specs)
    ax3.axis([-1,1,-1,1])
    ax3.set_zlim3d(-0.6, 0.6) 
    ax3.axis('off')
    title3 = "N=" + str(n) + "  k=" + str(k)
    ax3.set_title(title3, fontsize = 20)
    
    # SUBFIGURE 4
    # Create subfigure 4 
    n = 0; k = 3
    z4 = drumhead_height(n, k, rm, th, time)
    ax4 = fig.add_subplot(gs[2, 2], projection='3d')
    s4 = ax4.plot_surface(xm, ym, z4, **specs)
    ax4.axis([-1,1,-1,1])
    ax4.set_zlim3d(-0.6, 0.6) 
    ax4.axis('off')
    title4 = "N=" + str(n) + "  k=" + str(k)
    ax4.set_title(title4, fontsize = 20)
    
from IPython.html.widgets import interact
interact(plot_Bessel, time=(0,5,0.1))
```

```python
print("Frame number:")
nn = 5
for i in range(nn):
    time = i/100.0
    plot_Bessel(time)
    plt.savefig('animation_frames/bessel{0:07d}.png'.format(i+1));
    plt.close()
    print i,
```

```python
# Import JSAnimation 
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

# Choose location of files to use
filenames=sorted(glob.glob('animation_frames/bessel*.png'))

#Create figure
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=50, interval=20, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('animation_frames/anim_bessel.html', writer=HTMLWriter(embed_frames=True))

anim
```
