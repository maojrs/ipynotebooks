# Shallow water equations


The shallow water equations are used to describe the propagation of waves in fluid systems where the depth is much smaller than horizontal length scale. One dimensional shallow water equations are a nice example to understand non-linear wave propagation, for several reasons. It is a system of only two equations, and hence the simplest step up from the scalar case. Its non-linear structure is fairly simple, so it can be solved explicitly, and yet the structure is typical of what is seen in more complicated systems. It is physically very relevant for many realistic applications, like Tsunami modeling, and it is fairly easy to develop intuition on how its solutions will behave.

In order to derive the shallow water equations, consider the fluid is incompressible with constant density $\bar{\rho}$, velocity $u(x,t)$ and allow its depth $h(x,t)$ to vary. The total mass in $[x_1,x_2]$ at time $t$ is

$$
\int_{x_1}^{x_2} \bar{\rho} h(x,t)dx.
$$

The mass flux is given by $\bar{\rho}u(x,t)h(x,t)$; therefore, the conservation of mass yields,

$$
h_t + \left(uh\right)_x = 0
$$

The conservation of momentum follows the same form as in the Euler Equations,

$$
\left(\bar{\rho}hu\right)_t + \left(\bar{\rho}hu^2 + p\right)_x.
$$

However, in this case, the pressure is determined from a hydrostatic law, whcih means that the pressure at distance $h-y$ below the surface is $\bar{\rho}g(h-y)$, where $g$ is the gravitational constant. Integrating this from $y=0$ to $y=h(x,t$ gives the total pressure felt at $(x,t)$, $p=\frac{1}{2}\bar{\rho}$gh^2. Using this in the momentum euqation and canceling $\bar{\rho}$ yields,

$$
\left(hu\right)_t + \left(hu^2 + \frac{1}{2}gh^2\right)_x =0.
$$

Along with the conservation of masss, these two equations are the system of one-dimensional shallow water equations:

$$
 \left[ \begin{array}{c}
h \\
hu 
\end{array} \right]_t
+  \left[ \begin{array}{c}
uh \\
hu^2 + \frac{1}{2}gh^2  \\
\end{array} \right]_x= 0,
$$
 see [LeVeque 2002](http://depts.washington.edu/clawpack/book.html) for details.
 
The Riemann problem is the key ingredient for complex numerical simulations using shallow water equations and other hyperbolic systems. In this case, it consists of solving this system with the initial conditions:

$$
q(x,0) = \begin{cases}
\bar{q_l} & \text{if   } x \le 0, \\
\bar{q_r} & \text{if   } x > 0,
\end{cases}
$$

where $\bar{q} = [h, hu]$. This notebook can illustrate the solution to several Riemann problems
for this system. The main one dealt in this notebook is the dam break problem. However, the user can play with the values to see other solutions. The solution is showed as an animation of the evolution of the depth $h(x,t)$ along with a passive tracer (the color of the water) that helps visualize the fluid velocity.  

The solution always consists of two waves, each of which might be a shock wave or a rarefaction wave depending on where the left and right states lie in the phase plane.  An interactive phase plane plot is shown that illustrates how the intermediate state and the nature of each wave can be determined by looking at the intersection of Hugoniot loci and/or integral curves.  This theory is discussed in Chapter 13 of the book "Finite Volume Methods for Hyperbolic Problems" and gives more dynamic version of some of the plots in that chapter, see also <http://clawpack.github.io/doc/fvmbook.html>. 

The last section is devoted to a brief introduction to transverse Riemann problems, which are necessary to obtain second order accuracy in two dimensional problems. These can be implemented for shallow water equations to obatin accurate results in two dimensions.

See <http://clawpack.github.io/doc/notebooks.html> for other Clawpack notebooks.

## Dam break Riemann problem:

One very interesting Riemann problem for the sahllow water equations is the dam break problem. It can
model how will water behave when a dam breaks in idealized condition. Before the dam breaks, the initial velocity is zero and the depth is greater to the left of $x=0$ than to the right. As soon as the dam breaks, the water will begin moving. The solution will be shock wave moving to the right and a rarefaction wave moving to the left. The intermediate fluid velocity is positive. In the next animation, we show the solution of the dam break problem using Clawpack.

## The transverse Riemann problem in 2D

In order to obtain second order accuracy using unsplit methods, we will need to solve a transverse Riemann solver in addition to the normal one. This will require to split the normal wave fluctuactions $A^\pm \Delta Q_{i- 1/2,j}$ at edge $i-1/2$ into transverse wave fluctuactions $B^\pm A^+\Delta Q_{i- 1/2,j}$ and $B^\pm A^-\Delta Q_{i- 1/2,j}$; the former descomposition is shown in the following figure and explained in detail on [LeVeque 2002](http://depts.washington.edu/clawpack/book.html).
<img style="float: center" src="../images/transverse_diag.png", width=500>

We will only focuse on $B^\pm A^+\Delta Q_{i- 1/2,j}$, since the other case is analogous. We begin by decomposing the normal wave fluctuations into a linear combination of transverse waves, which will follow the same equations and jacobian as before, so we obtain

$$
A^+\Delta Q_{i- 1/2,j} = \mathbf{R_y} \bar{\beta} = \beta_D \bar{r_D} + \beta_M \bar{r_M} + \beta_U \bar{r_U},
$$

where the subindexes $B,M$ and $U$ denote bottom, middle and upper direction. Where $\bar{r_i}$ are the column eigenvectors of $\mathbf{B}$, if sweeping in the $x$ direction or of $\mathbf{A}$ if sweeping on the $y$ direction. In this case, we will only consider the case when sweeping in the $x$ direction, since the other is analogous.

Now we will show how the normal and transverse flucutuations propagate from the grid edges to the grid cells. We first
obtain the solution for the normal and transverse fluctuation and then we plot them. Note this plot doesn't really show the difference in the values for the depth and velocities, but it does illustrate the wave speeds and how the normal and transvere fluctuations propagate.
